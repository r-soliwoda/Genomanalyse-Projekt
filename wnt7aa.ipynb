```python
CDS=open('wnt7aa_CDS.fa').read()

fw=CDS[CDS.find("ATG"):CDS.find("ATG")+20] ##forward primer starts at ATG-position and goes on for 20 nucleotides
fw_tm=(2*(fw.count("A")+fw.count("T")))+(4*(fw.count("G")+fw.count("C"))) ##Tm calculated with formula from course: 2*(T+A)+4*(G+C)

x=0
y=0
if fw_tm>62:
    print("No suitable forward primer could be found due to Tm being already too high at 20 nucleotides")
    x=1


if fw_tm<55:
    lf=21 ##length-variable of forward primer for while loop
    while lf<24: ##primers need to be between 20-23 nucleotides long (so under 24)
        x=0 ##Testvariable ob break eingetreten ist
        fw=CDS[CDS.find("ATG"):CDS.find("ATG")+lf]
        fw_tm=(2*(fw.count("A")+fw.count("T")))+(4*(fw.count("G")+fw.count("C")))
        if fw_tm>=55 and fw_tm<=62: ##end loop if tm is now in range
            break
        lf=lf+1 ##length gets higher by 1 every "round" of the while loop
        x=1        
    if lf==25:
        print("No suitable forward primer could be found due to Tm not getting high enough in the range of 20-23 nucleotides of length")

rv=CDS[len(CDS)-20:] ##reverse primer for now equals the last 20 nucleotides of the CDS
rv_tm=(2*(rv.count("A")+rv.count("T")))+(4*(rv.count("G")+rv.count("C")))

if rv_tm>62:
    print("No suitable reverse primer could be found due to Tm being already too high at 20 nucleotides")
    y=1


if rv_tm<55:
    lr=21 ##length-variable of reverse primer for while loop
    while lr<24: ##primers need to be between 20-23 nucleotides long (so under 24)
        y=0 ##Testvariable (s.o.)
        rv=CDS[len(CDS)-lr:]
        rv_tm=(2*(rv.count("A")+rv.count("T")))+(4*(rv.count("G")+rv.count("C")))
        if rv_tm>=55 and rv_tm<=62: ##end loop if tm is now in range
            break
        lr=lr+1 ##length gets higher by 1 every "round" of the while loop
        y=1
    if lr==25:
        print("No suitable reverse primer could be found due to Tm not getting high enough in the range of 20-23 nucleotides of length")

z=0
q=0

if x==0 and y==0: ##there was a good forward and reverse primer found within the temperature range of 55-62°C
    diff_tm=abs(fw_tm-rv_tm) ##calculate difference between melting temperatures

    if diff_tm>4: ##if difference is too high (more than 4°C)
        if fw_tm>rv_tm: ##for the case that the melting temp. of the forward primer is higher
            lr=len(rv)+1 ##length-variable of reverse primer for while loop
            while lr<24: ##primers need to be between 20-23 nucleotides long (so under 24)
                q=0
                rv=CDS[len(CDS)-lr:]
                rv_tm=(2*(rv.count("A")+rv.count("T")))+(4*(rv.count("G")+rv.count("C")))
                diff_tm=abs(fw_tm-rv_tm)
                if diff_tm<=4: ##stop loop if difference of melting temperatures is now within range of 4°C
                    break
                lr=lr+1 ##length gets higher by 1 every "round" of the while loop
                q=1
    
        if rv_tm>fw_tm: ##if melting temp. of reverse primer is higher
            lf=len(fw)+1 ##length-variable of forward primer for while loop
            while lf<24: ##primers need to be between 20-23 nucleotides long (so under 24)
                z=0
                fw=CDS[CDS.find("ATG"):CDS.find("ATG")+lf]
                fw_tm=(2*(fw.count("A")+fw.count("T")))+(4*(fw.count("G")+fw.count("C")))
                diff_tm=abs(fw_tm-rv_tm)
                if diff_tm<=4:
                    break
                lf=lf+1 ##length gets higher by 1 every "round" of the while loop
                z=1
            
        if q==1 or z==1:
            print("No suitable primer pair with acceptable difference in Tm could be found")
            
        if q==0 or z==0:
            rv=rv.lower()
            rv=rv.replace("a", "T")
            rv=rv.replace("t", "A")
            rv=rv.replace("g", "C")
            rv=rv.replace("c", "G")
            rv=rv[::-1] ##rv replaced with itself, read from last position to first position, always taking 1 step backwards
            ##rv is now the reverse complement of the designed intermediate primer taken from the end of CDS -> rv is now a functioning reverse primer
            print("The forward primer is " + fw + " with a melting temperature of " + str(fw_tm) + "°C and the reverse primer is " + rv + " with a melting temperature of " + str(rv_tm) + "°C. The difference between the melting temperatures is " + str(diff_tm) + "°C.")
            
    elif diff_tm<=4 :
        rv=rv.lower()
        rv=rv.replace("a", "T")
        rv=rv.replace("t", "A")
        rv=rv.replace("g", "C")
        rv=rv.replace("c", "G")
        rv=rv[::-1] ##rv replaced with itself, read from last position to first position, always taking 1 step backwards
        ##rv is now the reverse complement of the designed intermediate primer taken from the end of CDS -> rv is now a functioning reverse primer
        print("The forward primer is " + fw + " with a melting temperature of " + str(fw_tm) + "°C and the reverse primer is " + rv + " with a melting temperature of " + str(rv_tm) + "°C. The difference between the melting temperatures is " + str(diff_tm) + "°C.")
        
        
if x==0 and y==1:
    print("The forward primer is " + fw + " with a melting temperature of " + str(fw_tm) + "°C.")
    
if x==1 and y==0:
    rv=rv.lower()
    rv=rv.replace("a", "T")
    rv=rv.replace("t", "A")
    rv=rv.replace("g", "C")
    rv=rv.replace("c", "G")
    rv=rv[::-1] ##rv replaced with itself, read from last position to first position, always taking 1 step backwards
    ##rv is now the reverse complement of the designed intermediate primer taken from the end of CDS -> rv is now a functioning reverse primer      
    print("The reverse primer is " + rv + " with a melting temperature of " + str(rv_tm) + "°C.")
```
